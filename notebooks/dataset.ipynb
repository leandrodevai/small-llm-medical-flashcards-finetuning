```python
import os

from datasets import DatasetDict, load_dataset
from dotenv import load_dotenv
from huggingface_hub import login

load_dotenv()

login(os.getenv("HF_TOKEN"))

SEED = 42
```

```python
dataset_complete = load_dataset("flwrlabs/medical-meadow-medical-flashcards")

dataset_complete
```

```python
dataset_train, dataset_eval = (
    dataset_complete["train"].train_test_split(train_size=30000, seed=SEED).values()
)
dataset_eval, dataset_test = dataset_eval.train_test_split(
    test_size=0.5, seed=SEED
).values()
```

```python
splitted_datasets = DatasetDict(
    {
        "train": dataset_train,
        "validation": dataset_eval,
        "test": dataset_test,
    }
)
splitted_datasets
```

```python
splitted_datasets.push_to_hub(
    "leandrodevai/medical-meadow-medical-flashcards-splits", private=False
)
```
